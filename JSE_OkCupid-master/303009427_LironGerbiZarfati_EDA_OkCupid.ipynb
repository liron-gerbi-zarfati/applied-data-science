I wanted to analyze if there is a connection between smoke and education or age. 

```python
edusmoke = df.groupby(by=['smokes','education']).count()['sex'].unstack()
edusmoke.head()
```

When I plottes the heatmap of smokes vs education,  I found out that there are windows in this information and it's not the most appropriate graph to display this. The windows are not necessarily missing information

```python
plt.figure(figsize=(12,6))
sns.heatmap(edusmoke,cmap='viridis')
```

We can see that most of the people who graduated from college/university aren't smoking. Or that most of the people who not smoking are graduated from university. (The yellow area).

The heatmap plot below of smokes vs age, also has missing information. We can understand some intresting things from this plot: 
1. There is a profile of a 109 year old man. This profile is probably fabricated
2. Most of the people in the data aren't smoking
3. Most of the users aged 24 to 29
4. Most of the young users aren't smoking.

```python
agesmoke = df.groupby(by=['smokes','age']).count()['sex'].unstack()
agesmoke.head()
plt.figure(figsize=(12,6))
sns.heatmap(agesmoke,cmap='viridis')
```

This is another plot of smoking information. Here you can see better that most users don't smoke.
